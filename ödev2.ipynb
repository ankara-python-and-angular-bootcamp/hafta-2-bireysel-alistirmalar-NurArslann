```python
#fibonacci
def fib(n):
    lst=[1,1]
    if n>2:
        for i in range(1,n+1):
            lst.append(lst[i]+lst[i-1])
        print(lst[:n])
    else:
        print(lst[:n])
fib(1)
```

```python
#Mükemmel sayılar
def mukemmel(n):
    sum=0
    for i in range(1,n):
        if(n%i == 0):
            sum +=i
        
    if(n == sum):
        return True
    else:
        return False
mukemmel(10)
```
